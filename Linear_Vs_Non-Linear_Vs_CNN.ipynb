```python
import pandas as pd
import matplotlib.pyplot as plt
import random
import torch
import torchvision
from torchvision import datasets
RANDOM_SEED = random.seed(42)
```

```python
# These data were taken from the saved models performances
model_0_results = {'model_name': 'FashionMNISTModelV0',
 'model_loss': 0.49280303545272386,
 'model_acc': 82.95726837060703,
 'training_time':131.91}

model_1_results = {'model_name': 'FashionMNISTModelV1',
 'model_loss': 0.4638106187644858,
 'model_acc': 83.30670926517571,
 'training_time' :4208.89
}

model_2_results = {'model_name': 'FashionMNISTModelV2',
 'model_loss': 0.35959244219544595,
 'model_acc': 85.97244408945687,
 'training_time':7663.90

 }
```

```python
compare_results = pd.DataFrame([model_0_results, model_1_results, model_2_results])
compare_results
```

```python
fig, ax = plt.subplots(figsize=(10, 2))
ax.axis('off')
table = ax.table(cellText=compare_results.values, colLabels=compare_results.columns, loc='center')
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.5)

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
sns.set(style="whitegrid")
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

sns.barplot(data=compare_results, x="model_name", y="model_acc", ax=axs[0], palette="Blues_d")
axs[0].set_title("Model Accuracy")
axs[0].set_ylabel("Accuracy (%)")
axs[0].set_xlabel("")

sns.barplot(data=compare_results, x="model_name", y="model_loss", ax=axs[1], palette="Reds_d")
axs[1].set_title("Model Loss")
axs[1].set_ylabel("Loss")
axs[1].set_xlabel("")

sns.barplot(data=compare_results, x="model_name", y="training_time", ax=axs[2], palette="Greens_d")
axs[2].set_title("Training Time")
axs[2].set_ylabel("Time (sec)")
axs[2].set_xlabel("")

plt.tight_layout()
plt.show()
```
